```python
#import libraries

import pandas as pd
import numpy as np
import seaborn as sns
from datetime import datetime


import matplotlib
import matplotlib.pyplot as plt
#import matplotlib.mlab as mlab

plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
                                       
#size of the plot                                       
matplotlib.rcParams['figure.figsize'] = (12,8)
                                       
#read data
                                       
df = pd.read_csv(r'C:\Users\tayos\OneDrive\Documents\movies.csv')
```

```python
df.head()
```

```python
#check for missing data
# loop through the data and see if there is anything missing
#format with col and percentage of missing data(null)

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
#data types for our columns

df.dtypes
```

```python
#get rid of .0 at the end of number in budget and gross column

#first replace non finite integers in budget column with NaN and convert them to a float data type

df['budget'] = df['budget'].replace([np.inf, -np.inf], np.nan).astype(float)

#replace NaN values in budget column with 0
df['budget'] = df['budget'].fillna(0)

#replace non finite integers in gross column with NaN and convert them to a float data type
df['gross'] = df['gross'].replace([np.inf, -np.inf], np.nan).astype(float)

#replace NaN values in gross column with 0
df['gross'] = df['gross'].fillna(0)

#convert gross and budget columns to int data type
df['gross'] = df['gross'].astype('int64')
df['budget'] = df['budget'].astype('int64')
```

```python
df.head()
```

```python
#making sure year column and released column have matching dates
#create yearcorrect column and fill it with the yyyy in release column



df['yearcorrect'] = df['released'].astype(str).apply(lambda x: x.split(' (')[0]).str[-4:]
#df['yearcorrect'] = df['released'].astype(str).str[-4:]
#df['yearcorrect'] = pd.to_datetime(df['yearcorrect'], format='%B %d, %Y').dt.year
#df.head()

```

```python
df.head()
```

```python
#order by gross revenue

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
#remove duplicates
df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
df.head()
```

```python
#scatter plot budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross earnings')

plt.xlabel('gross earning')

plt.ylabel('Budget for film')

plt.show()
```

```python
#plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df,scatter_kws={"color": "red"}, line_kws={"color":"blue"})
```

```python
#checking correlation, using different methods. pearson kendall spearman
#we find that theres a high sorrelation between budget and gross
df.corr(method='pearson')
```

```python
#visualise this data
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized.head()
    
```

```python
correlation_matrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation matric for numeric features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```
